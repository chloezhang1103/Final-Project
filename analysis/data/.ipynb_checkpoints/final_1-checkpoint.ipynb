```python
import pandas as pd

# 导入 CSV 文件
file_path = "pythonfinal/data/trip advisor restaurents  10k - trip_rest_neywork_1.csv"  # 替换为你的文件路径
df = pd.read_csv(file_path)

df.head()
```

```python
dataB = pd.read_csv("pythonfinal/data/DOHMH_New_York_City_Restaurant_Inspection_Results_20241213.csv")
```

```python
dataB.head()
```

```python
def clean_name(name):
    if pd.isnull(name):
        return ""
    return name.lower().strip().replace("’", "'").replace(".", "").replace(",", "")

# 标准化名称
df['cleaned_name'] = df['Title'].apply(clean_name)  # 替换为名称列名
dataB['cleaned_name'] = dataB['DBA'].apply(clean_name) 
```

```python
df.head()
```

```python
dataB.head()
```

```python
dataB = dataB.drop_duplicates(subset='cleaned_name', keep='first')
print(f"After removing duplicates, dataB size: {len(dataB)}")
```

```python
merged_data = pd.merge(df, dataB[['cleaned_name', 'Latitude', 'Longitude']], 
                       on='cleaned_name', how='left')
merged_data.head()
```

```python
print(f"After removing duplicates, dataB size: {len(merged_data)}")
```

```python
missing_location_count = merged_data['Latitude'].isnull().sum()

# 打印结果
print(f"Number of restaurants without geolocation: {missing_location_count}")
```

```python
merged_data_cleaned = merged_data.dropna(subset=['Latitude', 'Longitude'])
```

```python
print(f"After removing duplicates, dataB size: {len(merged_data_cleaned)}")
```

```python
merged_data_cleaned.head()
```

```python
import geopandas as gpd
from shapely.geometry import Point


merged_data_cleaned['geometry'] = merged_data_cleaned.apply(
    lambda row: Point(row['Longitude'], row['Latitude']), axis=1
)

geo_df = gpd.GeoDataFrame(merged_data_cleaned, geometry='geometry')

geo_df.set_crs(epsg=4326, inplace=True)

geo_df.head()
```

```python
# 定义合理的经纬度范围
latitude_min, latitude_max = 40.4774, 40.9176
longitude_min, longitude_max = -74.2591, -73.7002

# 筛选数据
geo_df_cleaned = geo_df[
    (geo_df['Latitude'] >= latitude_min) & (geo_df['Latitude'] <= latitude_max) &
    (geo_df['Longitude'] >= longitude_min) & (geo_df['Longitude'] <= longitude_max)
]

# 查看清理后的数据大小
print(f"Total records after removing outliers: {len(geo_df_cleaned)}")
```

```python
import matplotlib.pyplot as plt

# 绘制清理后的点
geo_df_cleaned.plot(marker='o', color='blue', markersize=5, figsize=(10, 6))

plt.title("NYC Restaurants (Cleaned)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid()
plt.show()
```

```python
import pandas as pd
from collections import Counter
import re
```

```python
print(merged_data_cleaned.columns)
```

```python
# 提取 review comment 列
merged_data_cleaned = merged_data_cleaned.rename(columns={'Reveiw Comment': 'Review Comment'})

```

```python
comments = merged_data_cleaned['Review Comment'].dropna()

# 合并所有评论为一个长字符串
all_text = " ".join(comments)

# 清理文本，准备分析
all_text_cleaned = re.sub(r"[^\w\s]", "", all_text.lower())
words = all_text_cleaned.split()

# 统计词频
word_counts = Counter(words)
most_common_words = word_counts.most_common(20)

print("Most common words:", most_common_words)
```

```python
import matplotlib.pyplot as plt

# 将词频结果转换为 DataFrame
word_freq_df = pd.DataFrame(most_common_words, columns=['Word', 'Frequency'])

# 绘制条形图
plt.figure(figsize=(10, 6))
plt.bar(word_freq_df['Word'], word_freq_df['Frequency'])
plt.xticks(rotation=45)
plt.title("Top 20 Most Frequent Words in Review Comments")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
dating_keywords = ['romantic', 'vibe', 'dating', 'love', 'date']

dating_related_counts = {word: word_counts[word] for word in dating_keywords}

print("Dating-related keywords and their counts:", dating_related_counts)
```

```python
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize

# 如果第一次使用 nltk，请下载必要的资源
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

# 对评论的所有文本进行分词
tokens = word_tokenize(all_text_cleaned)

# 进行词性标注
tagged_words = pos_tag(tokens)

# 过滤出形容词（词性以 'JJ' 开头的词）
adjectives = [word for word, tag in tagged_words if tag.startswith('JJ')]

# 统计形容词的出现频率
adjective_counts = Counter(adjectives)

# 显示出现频率最高的形容词
most_common_adjectives = adjective_counts.most_common(20)
print("Most common adjectives:", most_common_adjectives)
```

```python
dating_related_adjectives = ['romantic', 'cozy', 'intimate', 'lovely', 'beautiful', 'quiet', 'charming', 
                   'private', 'ambiance', 'special', 'memorable','love']

dating_adjective_counts = {word: adjective_counts[word] for word in dating_related_adjectives if word in adjective_counts}

print("Dating-related adjectives and their counts:", dating_adjective_counts)
```
